## Задача 1

Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалитетакие значения из списка и посчитайте суммарные затраты  
[100, 125, -90, 345, 655, -1, 0, 200]  
Используйте list comprehensions 

Пояснение:    
Дан список adv с затратами на рекламу.    
Но в данных есть ошибки, некоторые затраты имеют отрицательную величину.
Удалите такие значения из списка и посчитайте суммарные затраты.    
Запишите их в переменную x.
Используйте list comprehensions.    

Пример  
На входе:  
adv = [100, 125, -90, 345, 655, -1, 0, 200]  
  
На выходе:  
1425

```python
adv = [100, 125, -90, 345, 655, -1, 0, 200]
new_adv = []
for i in range(len(adv)):
    if adv[i] > 0:
        new_adv.append(adv[i])
new_adv

x = sum(new_adv)

# Другая запись:
# x = sum([i for i in adv if i > 0])

print(x)
```

На складе лежат разные фрукты в разном количестве.  
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе.

Пояснение:  
На складе лежат разные фрукты в разном количестве.  
Нужно написать функцию total_fruits, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе.  
  
можно решить через *kwargs  

Пример
После вашего кода будет добавлена следующая строка:  
print(total_fruits(apples=10, bananas=5, oranges=8))  

На выходе:

```python
def total_fruits(**kwargs):

    # Найдем сумму всех значений из входного словаря
    return sum(kwargs.values())

# Другая запись функции
def total_fruits(**fruits):
    summa = 0 
    # Перебираем все значения из словаря и через цикл складываем все значения
    for amount in fruits.values():
        summa += amount
    return summa    

# Создадим пустой список 
# count_fruits = []
# Поместим в пустой список значения из словаря, отвечающих за количество фруктов
# count_fruits.append(kwargs.values())
# print(count_fruits) #[dict_values([10, 5, 8])]

print(total_fruits(apples=10, bananas=5, oranges=8, tyijktg=8)) 
```
