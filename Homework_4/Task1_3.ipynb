### 1.2. Изучите стоимости недвижимости


```python
plt.figure(figsize=(10,4))
plt.hist(df_h['price'], bins=37)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость, млн.руб.')
plt.xticks(np.arange(0,8*1e6,0.25*1e6), rotation=45)
plt.ylabel('Количество')
plt.yticks(range(0,9000,500))
plt.grid() #Сетка

;
```

### 1.3. Изучите распределение квадратуры жилой

### 2.1. Изучите распределение домов от наличия вида на набережную

### 2.2. Изучите распределение этажей домов


```python
data = df_h['floors'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(data, labels=data.index, autopct='%0.2f%%')
plt.title('Количество этажей')

;
```

## Задача 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.  
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
# Матрица корреляции
corr_matrix = df_h.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

;
```

```python
sns.jointplot(x=df_h['price'], y=df_h['sqft_living'], kind='reg')

;
```

Вывод: Чем больше жилплощадь дома, тем выше его стоимость.

```python
sns.jointplot(x=df_h['price'], y=df_h['grade'], kind='reg')

;
```

Вывод: Чем выше качество конструкции и дизайна дома, тем выше его стоимость.

```python
sns.boxplot(x=df_h['price'], y=df_h['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Distribution of price by grade')

;
```

Вывод: Наибольшую стоимость имеют дома с оценкой конструкции и дизайна = 13.

```python
sns.boxplot(x=df_h['price'], y=df_h['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Distribution of price by floors')

;
```

Вывод: Наибольшую стоимость имеют дома с количеством этажей 2,5.

```python
sns.boxplot(x=df_h['price'], y=df_h['view'].astype('str'), whis=1.5)
plt.xlabel('price')
plt.ylabel('view')
plt.title('Distribution of price by view')

;
```

Вывод: Наибольшую стоимость имеют дома с оценкой вида = 4.
